```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
#1
ts = 0.0005
time = np.arange(0, 0.03, ts / 2)

x = np.sin(520 * np.pi * time + np.pi / 3.)
y = np.sin(280 * np.pi * time - np.pi / 3.)
z = np.sin(120 * np.pi * time + np.pi / 3.)


fig, axs = plt.subplots(3)

fig.suptitle("B")
axs[0].plot(time, x)
axs[1].plot(time, y)
axs[2].plot(time, z)

for ax in axs:
    ax.set(xlabel='time (s)', ylabel='Amplitude')

fig.show()

fr = 200

time = np.arange(0, 0.03, 1/fr)

x = np.sin(520 * np.pi * time + np.pi / 3.)
y = np.sin(280 * np.pi * time - np.pi / 3.)
z = np.sin(120 * np.pi * time + np.pi / 3.)

fig, axs = plt.subplots(3)

fig.suptitle("C")
axs[0].stem(time, x)
axs[0].plot(time, x)

axs[1].stem(time, y)
axs[1].plot(time, y)

axs[2].stem(time, z)
axs[2].plot(time, z)

for ax in axs:
    ax.set(xlabel='time (s)', ylabel='Amplitude')

fig.show()
```

```python
#2

fr = 400
time = np.arange(0, 1, 1/1600)
sin = np.sin(2 * np.pi * time * fr)

plt.title("A")
plt.plot(time[:math.floor(0.1 * len(time))], sin[:math.floor(0.1 * len(time))])
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.show()

fr = 800
time = np.arange(0, 3, 1 / (3 * fr))
sin = np.sin(2 * np.pi * time * fr)

plt.title("B")
plt.plot(time[:math.floor(0.01 * len(time))], sin[:math.floor(0.01 * len(time))])
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.show()

fr = 240
time = np.arange(0, 1, 1 / (2 * fr))
sawtooth = fr * np.mod(time, 1 / fr)


plt.title("C")
plt.plot(time[:math.floor(0.1 * len(time))], sawtooth[:math.floor(0.1 * len(time))])
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.show()


fr = 300
time = np.arange(0, 1, 1 / (12 * fr))
square = np.sign(np.sin(2 * np.pi * time * fr))

plt.title("D")
plt.plot(time[:math.floor(0.1 * len(time))], square[:math.floor(0.1 * len(time))])
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.show()


#e
image = np.random.rand(128, 128)
plt.imshow(image, cmap='gray')
plt.show()

#f

def myFunction():
    ret = np.ones((128, 128))
    np.subtract(ret, np.random.rand(128, 128))
    for i in range(128):
        for j in range(128):
            ret[i][j] = ret[i][j] -  random.random() * np.cos(2 * np.pi * (i + j) * random.random())

    return ret

plt.imshow(myFunction(), cmap='gray')
plt.show()

```

```python
'''
Daca frecventa de esantionare este de 2000 Hz,
din formula fs = 1 / T =>
=> T = 1 / fs, unde fs este frecventa de esantionare iar T este perioada de esantionare
T = 1 / 2000
'''

print(1 / 2000)
```

```python
'''
Tinand cont ca T = 1 / 2000, rezulta ca vor fi 3600 / T esantioane intr-o ora
#esantioane = 3600 / (1 / 2000) = 3600 * 2000 = 7.2 * 10^6
'''

nr_esantioane = 3600 / (1 / 2000)
print(nr_esantioane)

'''
Un byte este format din 8 biti, iar cum un esantion este memorat pe 4 biti, rezulta ca un esantion
ocupa 1/2 bytes.
#bytes = 7.2 * 10^6 * 1/2 = 3.6 * 10^6 bytes = '''

print(nr_esantioane / 2)
```
