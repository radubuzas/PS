```python
#3
import pandas as pd

df = pd.read_csv('./airQuality.csv')

df = df.drop('decimal', axis=1)

last = df.iloc[0]['mon']

months = []

s, cnt = 0, 0
for _, row in df.iterrows():
    if row['mon'] != last:
        last = row['mon']
        months.append(s / cnt)
        s, cnt = 0, 0
    else:
        s += row['ppm']
        cnt += 1

plt.plot(months)
plt.show()
```
